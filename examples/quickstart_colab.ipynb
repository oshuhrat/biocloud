```
from biocloud import BioCloud

sim = BioCloud('MCF7', grid_size=80)
```

```
# I-V curve and bistability
sim.plot_iv()
```

```
# Critical cloud size
sim.critical_size_scan()
```

```
# Phase diagram (~10 min)
sim.phase_diagram()
```

```
# Instructor cell effect
sim.instructor_comparison(cloud_radius=8)
```

```
# Therapy comparison
sim.therapy_comparison(cloud_radius=8)
```

```
# Summary
sim.summary()
```
